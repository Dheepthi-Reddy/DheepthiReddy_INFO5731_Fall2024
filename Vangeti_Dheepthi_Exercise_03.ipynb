## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):

import requests                   # 'requests' library imported to send HTTP requests to fetch web pages
from bs4 import BeautifulSoup     # 'BeautifulSoup' library imported to parse HTML content and extract data and extracting data from web pages
import csv                        # 'csv' library imported to read and write in CSV files
import pandas as pd               # 'pandas' library for data manipulation operatioins
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer    # necessary tools from 'sklearn' for converting text to numerical feature vectors.
import nltk                       # 'nltk' Natural Language Tool Kit library for natural language processing
from nltk.sentiment import SentimentIntensityAnalyzer    # 'SentimentIntensityAnalyzer' from 'nltk' for sentiment analysis
from nltk.util import ngrams      # a library from nltk to create n-grams from text


# to avoid blocking from web pages while scraping we use headers to create a client request like a genuine web page
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

# 'getImdbReviews' function defined to scrape the reviews from a specified url
def getImdbReviews(url, num_pgs=200):  # the function has 2 arguments, url to the web page and the number of pages to be scraped

    # initialised a empty list to store reviews
    reviewsList = []
    # to loop through each page
    for pg in range(1, num_pgs + 1):
        # a get request is sent to the web page to get HTML content
        response = requests.get(url + '?start=' + str((pg-1)*50), headers=headers)
        # parsing html content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        # to find all th conent which has div class name as text show-more__control
        reviewDiv = soup.find_all('div', {'class': 'text show-more__control'})
        # for loop to append each review to the empty reviewsList
        for review in reviewDiv:
            reviewsList.append(review.text.strip())
        # if no more reviews found then we break the loop
        if not reviewDiv:
          break
    # returns the list of reviews
    return reviewsList

# 'saveToCsv' function defined for saving these scraped reviews to a CSV file, it has the reviews list and filename as its parameters
def saveToCsv(reviews, filename="Reviews.csv"):

    # opens the csv file in write mode
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        # writes the header row with two column names mentioned below
        writer.writerow(["Review Number", "Review Text"])
        # loops through each review whiile writing them to the csv file
        for i, review in enumerate(reviews, 1):
            writer.writerow([i, review])

# input url of IMDb movie review page
url = 'https://www.imdb.com/title/tt6263850/reviews/?ref_=ttrt_sa_3'
# applying function to fetch reviews from the url
reviews = getImdbReviews(url)
# saving first 100 reviews to csv file
saveToCsv(reviews[:1000])

print("Extracted reviews text data is saved to Reviews.csv")

# Creating a DataFrame to save the reviews
df = pd.DataFrame(reviews, columns=['Review Text'])

# to print first 10 rows of the dataframe
df.head(10)
```

```python
# nltk library downloads the VADER lexicon
nltk.download('vader_lexicon')
# initializing the SentimentIntensityAnalyzer, which is used to calculate sentiment scores.
intensityAnalyzer = SentimentIntensityAnalyzer()

# defining 'sentimentScore' function to compute the sentiment score using VADER
def sentimentScore(text):
    return intensityAnalyzer.polarity_scores(text)['compound']
# defining 'bigrams' function for extracting bigrams from the text
def bigrams(text):
    return list(ngrams(text.split(), 2))
# defining 'reviewLength' function for calculating the number of words
def reviewLength(text):
    return len(text.split())
# defining 'wordCount' function for counting the total number of words
def wordCount(text):
    return len(text.split())
# defining 'averageWordLength' function for calculating the average length of the words
def averageWordLength(text):
    words = text.split()
    return sum(len(word) for word in words) / len(words) if words else 0

# creating a DataFrame to store these reviews
df = pd.DataFrame(reviews, columns=['Review Text'])

# adding the sentiment scores to the DataFrame
df['Sentiment Score'] = df['Review Text'].apply(sentimentScore)
# adding extracted bigrams to the DataFrame
df['Bigrams'] = df['Review Text'].apply(bigrams)
# adding Review Length to the DataFrame
df['Review Length'] = df['Review Text'].apply(reviewLength)
# adding Review Length to the DataFrame
df['Word Count'] = df['Review Text'].apply(wordCount)
# adding Average Word Length to the DataFrame
df['Average Word Length'] = df['Review Text'].apply(averageWordLength)

# print the final DataFrame
print("Features:", df)

# save the DataFrame to a csv file
df.to_csv("FeaturesOfReviews.csv", index=False, encoding='utf-8')
print("Extracted Features are saved to FeaturesOfReviews.csv")
```

```python
# You code here (Please add comments in the code):


import pandas as pd         # Importing pandas for data manipulation
import numpy as np          # Importing numpy for numerical calculations
from sklearn.metrics.pairwise import cosine_similarity      # 'cosine_similarity' from sklearn for calculating cosine similarity between vectors
from transformers import BertTokenizer, BertModel      # 'BertTokenizer' and 'BertModel' from 'transformers' for using pretrained BERT model
import torch                # Importing torch for tensor operations

# loading the BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
# 'bert-base-uncased' is to not distinguish between uppercase and lowercase letters
model = BertModel.from_pretrained('bert-base-uncased')

# defining 'getBertEmbeddings' function to get BERT embeddings for a list of texts
def getBertEmbeddings(txt):
    # tokenizing the texts by tokenizing, padding, and truncating it to a maximum length of 512 tokens, to prepare as inputs to the BERT
    inputs = tokenizer(txt, padding=True, truncation=True, return_tensors='pt', max_length=512)
    # to generate input text without calculating any gradients
    with torch.no_grad():
        outputs = model(**inputs)
    # returning the mean to give a single embedding for each input text
    return outputs.last_hidden_state.mean(dim=1).numpy()

# defined a sample query
sampleQuery = "Amazing movie with great performances"
# 'df' with reviews of the movie
reviews_df = df['Review Text'].tolist()

# calculating the BERT embeddings for both the query and the reviews
sampleQuery_embedding = getBertEmbeddings([sampleQuery])
review_embeddings = getBertEmbeddings(reviews_df)

# finding cosine similarity score using query and reviews embeddings
cosineSimilarity = cosine_similarity(sampleQuery_embedding, review_embeddings)[0]

# adding a new column to the DataFrame for similarity scores
df['Similarity Score'] = cosineSimilarity

# droppping if any duplicates exist
df = df.drop_duplicates(subset='Review Text').reset_index(drop=True)

# sorting DataFrame by similarity score in descending order
similarity_df = df.sort_values(by='Similarity Score', ascending=False)
print(similarity_df[['Review Text', 'Similarity Score']].head(10))

# saving the ranked DataFrame to a CSV file
similarity_df.to_csv("RankedReviewsBasedOnSimilarity.csv", index=False, encoding='utf-8')
print("Ranked reviews based on text similarity saved to RankedReviewsBasedOnSimilarity.csv")
```
