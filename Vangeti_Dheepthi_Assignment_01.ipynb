# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Your code here
import math

print("r    V")

# for loop to iterate over the even numbers between 2 and 21,
for r in range(2, 21, 2):
  V = (2 * math.pi * (r ** 3))/3
  print(f"{r}    {V:.2f}")
```

# Question 2 (20 points)

(a) Write a Python program to generate a sequence of 12 random numbers falling within the range of 50 to 150.

(b) Arrange the generated numbers in ascending order. Display the sorted list.

(c) Extend the program to arrange the numbers in descending order as well. Present the sorted list accordingly.

(d) Compute and output the average of the entire group of generated numbers.

(e) Enhance the program to calculate and display the average of the first 6 numbers from both the ascending and descending sorted lists.

```python
# Write code for each of the sub parts with proper comments.

# (a)

import random

# using randint math function to generate the random numbers
nums = [random.randint(50,150) for i in range(12)]
print("Generated 12 random numbers:", nums)

# (b)

# predeined sorted function used to sort the random numbers in Ascending order
ascending_num = sorted(nums)
print("Generated Random numbers in Ascending order: ", ascending_num)

# (c)

# doing reverse of ascending order list to get the descending order list
descending_num = sorted(nums, reverse = True)
print("Generated Random numbers in Descending order: ", descending_num)

# (d)

# sum of random numbers
sumOfNum = sum(nums)

# length of the random numbers
lenOfNum = len(nums)

# average of the list
average = sumOfNum / lenOfNum

print("Average of the random number list: ", average)

# (e)

# taking a subset of 6 elements from ascending order list
ascending_subset = ascending_num[:6]

# average of first 6 elements
ascending_avg = sum(ascending_subset) / len(ascending_subset)
print("First 6 elements average when arranged in Ascending order: ", ascending_avg)

# taking a subset of 6 elements from descending order list
descending_subset = descending_num[:6]

# average of first 6 elements
descending_avg = sum(descending_subset) / len(descending_subset)
print("First 6 elements when average arranged in Descending order: ", descending_avg)
```

# Question 3 (20 points).

Create a Python program that encompasses various string manipulation task:

(a) Prompt the user to input two sentences.

(b) Transform each sentence into a list, where each element corresponds to a word in the sentence. For example, if the sentences are "Hello, how are you doing today?" and "This is an example sentence for the program!"

The program should output:
['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(c) Compare the two sentences. Print out a list of words that occur in both sentences.

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
# Your code here

print("(a)")

# taking the sentences to input
str1 = input("Enter sentences 1 of 2: ")
str2 = input("Enter sentences 2 of 2: ")

print("\n(b)")

# importing regular expressions function
import re

def sentenceToList(sentence):
  # for seperating words based on non-alphanumeric charecters
  li = re.findall(r"[\w']+|[.,!?;]", sentence)
  return li

finalstr = str1 + str2
print(sentenceToList(finalstr))

print("\n(c)")

# defining a function to find the words common in two sentences
def commonWords(sent1, sent2):

  # getting the words of each sentence
  str1_words = sentenceToList(str1)
  str2_words = sentenceToList(str2)

  # considering words of each sentence to a set and getting the intersection(coomon) words of the two sentences
  return sorted(set(str1_words).intersection(set(str2_words)))

print("Words common in both sentences: ", commonWords(str1,str2))

print("\n(d)")

# defining a sanitizedSent function to return a sentence without different charecters
def sanitizedSent(sentence):

  # using regular expressions function to specify these charecters
  regex = r"(?<!\d)[.,;:](?!\d)"

  # ".sub()" function of regular expressions is used for removing or replacing specific patterns
  return re.sub(regex, "", sentence, 0)

print("Sanitized sentence of the original sentence: ", sanitizedSent(str1))

print("\n(e)")

# defined a function to count the number of vowels
def vowelCount(sentence):

    # created a dictionary to initialize the initial count
    vowels = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}

    # for loop to iterate over each charecter in the sentence
    for char in sentence.lower():
        if char in vowels:
            vowels[char] += 1
    return vowels

print("Sentence1: ", vowelCount(str1))
print("Sentence2: ", vowelCount(str2))
```

# Question 4 (20 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
import re # regular expressions
import math #loading math library

# Error handling for Invalid input expression
class InvalidExpressionError(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)
print(str)

# Checking if the given string is a valid expression or not
def checkInputExpression(inputExp):
    # validating the input mathematical expression
    # desiredExp = re.compile(r'[0-9+\-*/().\s]*')
    desiredExp = re.compile(r'^[0-9+\-*/().\s%]*\s*sqrt\([0-9+\-*/.\s]+\)\s*$|^[0-9+\-*/().\s%]*$')
    # matching the entire input expression with the desired expression format
    mathExp = desiredExp.fullmatch(inputExp)
    # print(mathExp)
    if mathExp:
      return None
    else:
      # rasising the invalid expression error when the input expression format is wrong
      raise InvalidExpressionError("Error: Invalid input expression")


# checking for parenthesis in the input expression
def checkForParanthesis(inputExp):
    # print("Running paranthesis block")
    while '(' in inputExp:
        # check for opening parenthesis
        start = inputExp.rfind('(')
        # check for the closing parathesis
        end = inputExp.find(')', start)

        # checking if all the opening parathesis has a corresponding closing parenthesis
        if start == -1 or end == -1:
            raise ValueError("Mismatched parentheses")
        # Calculate the result within the parenthesis
        sub_expr = inputExp[start + 1:end]
        result = calc(sub_expr)
        # Replace the parenthesis with the result
        inputExp = inputExp[:start] + str(result) + inputExp[end + 1:]
        print("First stage "+inputExp)
    return inputExp

# checking for ** in the input expression
def checkForExponents(inputExp):
    # print("Running Exponents block to check for expo")
    while '**' in inputExp:
        # Find the position of exponentiation operator
        regexCheck = re.compile(r'(\d+\.?\d*)\s*\*\*\s*(\d+\.?\d*)')
        # checking the input pattern with defined pattern
        match = regexCheck.search(inputExp)
        base = float(match.group(1))
        exponent = float(match.group(2))
        result = math.pow(base, exponent)
        inputExp = inputExp[:match.start()] + str(result) + inputExp[match.end():]
        print("Second stage "+inputExp)
    return inputExp

# checking for Multiplication and Division in the input expression
def checkForMultiplyAndDivision(inputExp):
    # print("Running mutliplication and division block")
    regexCheck = re.compile(r'(\d+\.?\d*)\s*([*/])\s*(\d+\.?\d*)')
    while regexCheck.search(inputExp):
        # checking the input pattern with defined pattern
        match = regexCheck.search(inputExp)
        left = float(match.group(1))
        op = match.group(2)
        right = float(match.group(3))
        if op == '*':
           result = left * right
        elif op == '/':
          if right == 0:
            raise ZeroDivisionError("Division by Zero error")
          result = left / right
        inputExp = inputExp[:match.start()] + str(result) + inputExp[match.end():]
    return inputExp

# checking for Addition and Subtraction in the input expression
def checkForAdditionAndSubstraction(inputExp):
    # print("Running additon block")
    regexCheck = re.compile(r'(\d+\.?\d*)\s*([+-])\s*(\d+\.?\d*)')
    while regexCheck.search(inputExp):
          # checking the input pattern with defined pattern
          match = regexCheck.search(inputExp)
          # coverting base numbers to a float type
          left = float(match.group(1))
          op = match.group(2)
          right = float(match.group(3))

          # checking for the operation to perform
          if op == '+':
            result = left + right
          elif op == '-':
            result = left - right
          # performing the operation by taking the expression as start, result and end
          inputExp = inputExp[:match.start()] + str(result)+ inputExp[match.end():]
    return inputExp

#  Function to process square root and modulus
def checkForSqrtAndModulus(inputExp):
    # print("Running square root and modulus block")
    # Handle square root
    while 'sqrt(' in inputExp:
        regex_check = re.compile(r'sqrt\((\d+\.?\d*)\)')
        # checking the input pattern with defined pattern with sqrt()
        match = regex_check.search(inputExp)
        value = float(match.group(1))
        result = math.sqrt(value)
        # performing the opeartion on the input expresion
        inputExp = inputExp[:match.start()] + str(result) + inputExp[match.end():]
        print("After processing square root:", inputExp)
    # Handle modulus
    while '%' in inputExp:
        regex_check = re.compile(r'(\d+\.?\d*)\s*%\s*(\d+\.?\d*)')
        # checking the input pattern with defined pattern with '%' operator
        match = regex_check.search(inputExp)
        left = float(match.group(1))
        right = float(match.group(2))
        result = left % right
        # performing the opeartion on the input expresion
        inputExp = inputExp[:match.start()] + str(result) + inputExp[match.end():]
        print("After processing modulus:", inputExp)
    return inputExp

# Function to execute calculator
def calc(inputExp):
    """calculate the given string"""
    checkInputExpression(inputExp)                        # checking if the expression is in the right format to perform calculations
    inputExp = checkForSqrtAndModulus(inputExp)           # checking for square root and modulus calculation
    inputExp = checkForParanthesis(inputExp)              # checking for parathesis
    inputExp = checkForExponents(inputExp)                # checking for exponents
    inputExp = checkForMultiplyAndDivision(inputExp)      # checking for multiplication and division
    result = checkForAdditionAndSubstraction(inputExp)    # checking for addition and subtraction operations
    return result

print("Welcome to Python Calculator")
print('''
    - Valid format of inputs '1 + 2', '1.2 * 2.3', '2*(3-1)', '2**3'.
    - Operations supported: +, -, *, /, '**' to perform exponentials, 'sqrt()' to perform square root, '%' to perform modulus.
    - Enter 'quit' in the input box to stop the calculator operations''');

while True:
    # to take the user input and to remove any unwanted spaces using strip
    exp = input("Enter an operation to perform: ").strip()

    # to exit from the calculator operations
    if exp.lower() == 'quit':
        print("Exiting from the calculator....")
        break
    # to handle exceptions while performing calculator operations
    try:
        result = calc(exp)
        print(f"Result: {result}")
    except Exception as e:
        print(f"Error: {e}")
```

# Question 5 (30 points).

In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

To answer this question, you will need to download the file [Assignment1_denton_housing.csv](https://raw.githubusercontent.com/unt-iialab/info5731-spring2022/main/assignments/Assignment1_denton_housing.csv) included with this assignment. The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (10 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (10 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
 # Your code here

import pandas as pd
import numpy as np

print("(a)")

# reading the csv file with the help of python pandas
denton_housing = pd.read_csv("Assignment1_denton_housing.csv")

# filtering the 'title_field' column for 'Occupied Housing Units'
occupied = denton_housing[denton_housing['title_field'] == 'Occupied Housing Units']

# sorting with respect to years
occupied = occupied.sort_values(by = 'year')

occupied.head(10)

# sorting the values of occupied units
noOfUnits = occupied['value'].sort_values()

# sorting the differencing in ascending order
diff = np.sort(np.diff(noOfUnits))

print("The difference of Occupied Housing Units values: ", diff)

print("\n(b)")

# filtering the rows for the required fields
totalHousingUnits = denton_housing[denton_housing['title_field'] == 'Total Housing Units']
vacantHousingUnits = denton_housing[denton_housing['title_field'] == 'Vacant Housing Units']
occupiedHousingUnits = denton_housing[denton_housing['title_field'] == 'Occupied Housing Units']

# merging 3 filtered rows with respect to year
denton_housing_merged = pd.merge(totalHousingUnits, vacantHousingUnits, on = 'year')
denton_housing_merged = pd.merge(denton_housing_merged, occupiedHousingUnits, on = 'year')

# renaming the formed default column names
denton_housing_merged.rename(columns={'title_field_x' : 'title_field', 'title_field_y' : 'title_field', 'value_x' : 'total', 'value_y' : 'vacant', 'value' : 'occupied'}, inplace = True)

# percentages of vacant housings for all the years
denton_housing_merged['vacantPercentage'] = (denton_housing_merged['vacant'] / denton_housing_merged['total']) * 100
denton_housing_merged['occupiedPercentage'] = (denton_housing_merged['occupied'] / denton_housing_merged['total']) * 100

# sorting the results by year
denton_housing_merged = denton_housing_merged.sort_values(by='year')

# printing the year wise vacant and occupied housing percentages with the help of for loop
print("Year wise vacant and occupied units percentages:")
for i, row in denton_housing_merged.iterrows():
    year = row['year']
    vacant_percentage = f"{row['vacantPercentage']:.0f}%"
    occupied_percentage = f"{row['occupiedPercentage']:.0f}%"
    print(f"{year}  {vacant_percentage}  {occupied_percentage}")

print("\n(c)")

# using idmax to find the row with highest number of vacant and occupied housing
highestVacant = denton_housing_merged.loc[denton_housing_merged['vacant'].idxmax()]
highestOccupied = denton_housing_merged.loc[denton_housing_merged['occupied'].idxmax()]

# storing the results to print
results = pd.DataFrame({
    ' ': ['Vacant Housing Units  ', 'Occupied Housing Units'],
    'Year': [highestVacant['year'], highestOccupied['year']],
    'Value': [highestVacant['vacant'], highestOccupied['occupied']]
})

# printing the results
print(results.to_string(index = False))

# denton_housing_merged

print("\n(d)")

# stats library to find hmean
from scipy import stats

# harmonic mean of total housing units formula
hmeanOfTotalUnits = stats.hmean(denton_housing_merged['total'])

print(f"Harmonic mean of total housing units is: {hmeanOfTotalUnits:.3f}")
```
